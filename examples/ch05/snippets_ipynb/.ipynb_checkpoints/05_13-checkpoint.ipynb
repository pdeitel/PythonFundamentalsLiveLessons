# 5.13 Generator Expressions

```python
numbers = [10, 3, 7, 1, 9, 4, 2, 8, 5, 6]
```

```python
for value in (x ** 2 for x in numbers if x % 2 != 0):
    print(value, end='  ')
```

```python
squares_of_odds = (x ** 2 for x in numbers if x % 2 != 0)
```

```python
squares_of_odds 
```
