**3. _(IPython Session)_** String dictionary keys are case sensitive. Confirm this by using the following dictionary and assigning `10` to the key `'x'`—doing so adds a new key–value pair rather than correcting the value for the key `'X'`: 
```python
roman_numerals = {'I': 1, 'II': 2, 'III': 3, 'V': 5, 'X': 100}
```

**Answer:** 

````python
roman_numerals = {'I': 1, 'II': 2, 'III': 3, 'V': 5, 'X': 100}
````

````python
roman_numerals['x'] = 10
````

````python
roman_numerals
````
